Как видно из работы функции boxplot(), в данных присутствуют выбросы

Найдем и удалим эти выбросы для каждого признака:

```python
# определяем столбцы, в которых нужно найти выбросы
columns_to_check = ['BirthAsphyxia', 'HypDistrib', 'CO2', 'Grunting',
       'LVHreport', 'CO2Report', 'GruntingReport', 'LungParench',]

# находим квартили для каждого столбца
q1 = data[columns_to_check].quantile(0.25)
q3 = data[columns_to_check].quantile(0.75)

# находим межквартильный размах для каждого столбца
iqr = q3 - q1

# определяем границы выбросов для каждого столбца
lower_bound = q1 - 1.5 * iqr
upper_bound = q3 + 1.5 * iqr

# выводим выбросы для каждого столбца
for col in columns_to_check:
    outliers = data[(data[col] < lower_bound[col]) | (data[col] > upper_bound[col])]
    print("Выбросы в столбце", col)
    print(outliers)

# заменяем выбросы на квантиль для каждого столбца
for col in columns_to_check:
    data.loc[data[col] < lower_bound[col], col] = q1[col]
    data.loc[data[col] > upper_bound[col], col] = q1[col]
```

```python
# построение ящика с усами
fig, ax = plt.subplots(figsize = (30, 10))
ax.boxplot([data['BirthAsphyxia'], data['HypDistrib'], data['HypoxiaInO2'], data['CO2'],
           data['ChestXray'], data['Grunting'], data['LVHreport'], data['LowerBodyO2'],
           data['RUQO2'], data['CO2Report'], data['XrayReport'], data['Disease'],
           data['GruntingReport'], data['Age'], data['DuctFlow'],
           data['CardiacMixing'], data['LungParench'], data['LungFlow'], data['Sick']])
ax.set_xticklabels(['BirthAsphyxia', 'HypDistrib', 'HypoxiaInO2', 'CO2', 'ChestXray', 'Grunting',
       'LVHreport', 'LowerBodyO2', 'RUQO2', 'CO2Report', 'XrayReport', 'Disease',
       'GruntingReport', 'Age', 'DuctFlow', 'CardiacMixing', 'LungParench',
       'LungFlow', 'Sick'])
plt.show()
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score, confusion_matrix

# Создаем объекты моделей
logreg = LogisticRegression()
rf = RandomForestClassifier()
mlp = MLPClassifier(hidden_layer_sizes=[50, 100])

# Обучаем модели на обучающей выборке
logreg.fit(X_train, y_train)
rf.fit(X_train, y_train)
mlp.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix

# Получаем предсказания на тестовой выборке
y_pred_logreg = logreg.predict(X_test)
y_pred_rf = rf.predict(X_test)
y_pred_mlp = mlp.predict(X_test)

# Вычисляем accuracy и confusion matrix
accuracy_logreg = accuracy_score(y_test, y_pred_logreg)
cm_logreg = confusion_matrix(y_test, y_pred_logreg)

accuracy_rf = accuracy_score(y_test, y_pred_rf)
cm_rf = confusion_matrix(y_test, y_pred_rf)

accuracy_mlp = accuracy_score(y_test, y_pred_mlp)
cm_mlp = confusion_matrix(y_test, y_pred_mlp)

print("Accuracy логистической регрессии:", accuracy_logreg)
print("Confusion matrix логистической регрессии:\n", cm_logreg)

print("Accuracy RandomForest:", accuracy_rf)
print("Confusion matrix RandomForest:\n", cm_rf)

print("Accuracy полносвязной нейронной сети:", accuracy_mlp)
print("Confusion matrix полносвязной нейронной сети:\n", cm_mlp)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV

# задаем диапазон значений для гиперпараметров
param_grid = {
    'n_estimators': [50, 100, 200, 400, 800],
}

# создаем экземпляр модели RandomForest
rf = RandomForestClassifier()

# создаем экземпляр GridSearchCV
gs = GridSearchCV(rf, param_grid, cv=5)

# обучаем модель на тренировочных данных
gs.fit(X_train, y_train)

# выводим лучшие значения гиперпараметров и метрику качества
print('Best parameters:', gs.best_params_)
print('Best score:', gs.best_score_)
```

```python
from sklearn.metrics import confusion_matrix

# получаем предсказания модели на тестовых данных
y_pred = gs.predict(X_test)

# строим confusion matrix
cm = confusion_matrix(y_test, y_pred)
print(cm)
```

```python
cm = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', cbar=False,
            xticklabels=['Healthy', 'Sick'],
            yticklabels=['Healthy', 'Sick'])
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()
```

```python
import matplotlib.pyplot as plt

errors = []

cv_results = gs.cv_results_

for i, n in enumerate(param_grid['n_estimators']):
    errors.append(1 - cv_results['mean_test_score'][i])

plt.plot(param_grid['n_estimators'], errors, marker='o')
plt.xlabel('Number of Estimators')
plt.ylabel('Error')
plt.title('Error vs Number of Estimators')
plt.show()
```

Таким образом, логистичееская регрессия смогла правильно предсказать 76,96% всех объектов. Метод RandomForest смог правильно правильно предсказать 73,65% всех объектов. Полносвязная нейронная сеть с двумя скрытыми слоями смогла предсказать 76,85% всех объектов. Исходя из этого, можно сделать вывод, что все модели приблизительно одинаково эффективны.
